```python
import chromadb

# Initialize the client (ensure the path matches your MedCare setup)
client = chromadb.PersistentClient(path="./medcare_vector_db")

# Access the collection
collection = client.get_collection(name="medcare_internal_docs")

# Example: Searching for information about a specific medical condition
results = collection.query(
    query_texts=["Is there a way for me to talk to a doctor over the internet?"],
    n_results=2, # Return the top 2 most relevant chunks
)

# Print the found text
for i, text in enumerate(results['documents'][0]):
    print(f"Result {i+1}: {text}\n")
```
