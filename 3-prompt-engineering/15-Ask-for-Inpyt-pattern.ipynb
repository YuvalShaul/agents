## The Ask For Input pattern

The "Ask for Input" prompt pattern is a fundamental technique in prompt engineering designed to prevent the AI from generating a complete response prematurely. Instead of guessing your requirements or providing a generic answer, the AI is instructed to wait and ask you for specific information first.

It transforms the interaction from a monologue (where the AI guesses) into a dialogue (where the AI gathers requirements).


### How It Works
In a standard prompt, you give instructions and hope for the best.  
In the "Ask for Input" pattern, you explicitly tell the AI to pause.  
The structure usually follows these steps:
- Define the Goal: Tell the AI what you ultimately want to achieve.
- Declare the Constraint: Tell the AI not to generate the final result yet.
- Specify the Information Needed: List the categories of data you want it to collect from you.
- Initiate the Questioning: Direct the AI to start by asking the first question.

Comparison Table
| Feature | Standard Prompting  | "Ask for Input" Pattern | 
| Output Type | Direct answer/output | Guided questionnaire | 
| Accuracy | Varies (often contains "hallucinations" or assumptions)| High (grounded in specific user data)
| Effort | Low initial effort, high editing effort | Higher initial setup, low editing effort | 
| Control | AI-driven | User-driven|



#### Example 1

**Prompt:**
```
I want you to act as a professional copywriter. I want you to write a LinkedIn bio for me, but I do not want you to write it yet.
Before you begin, you must ask me a series of questions to understand my background, my target audience, and my tone of voice. Ask me one question at a time. Once you have enough information, tell me you are ready to write the bio.
Ask me your first question.
```
