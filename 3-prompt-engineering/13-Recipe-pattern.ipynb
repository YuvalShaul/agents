## The Recipe pattern

The Recipe Prompt Pattern is a structural technique used to guide an AI through a complex task by breaking it down into a predefined set of "ingredients" (inputs) and "instructions" (steps).

Think of it as providing a template where the AI knows exactly what information to look for and what sequence of actions to perform to reach a specific "dish" or outcome.

### The Core Components

A well-structured recipe prompt usually follows this anatomy:  
|  Component  | Description |  Example |
| The Goal    | A clear statement of what the final output should be. | "Generate a 500-word blog post."|
| Ingredients | The specific data points or variables the AI must use.| "Topic, Target Audience, Tone, Keywords."|
| The Process | A step-by-step sequence of operations.  | "1. Outline, 2. Draft, 3. Review for SEO."|
| Constraints | The "dos and don'ts" or formatting rules. | "No passive voice; use Markdown headings."|


### Why Use This Pattern?
Using a recipe pattern moves you away from "hoping" the AI gets it right and toward repeatable excellence.
- Consistency: You get the same style and quality of output every time you swap the ingredients.
- Modularity: You can easily update one "step" in the process without rewriting the entire prompt.
- Error Reduction: By defining the ingredients upfront, the AI is less likely to hallucinate or skip crucial details.

Becoming a sculptor is a deeply rewarding journey that blends technical mastery with creative vision. Since you are looking for a complete sequence of steps to go from a beginner to a professional with your own workshop, here is a structured roadmap.

---

## 1. Education and Skill Acquisition

Before opening a shop, you must master the physics of your chosen medium.

* **Choose Your Medium:** Decide whether you want to work with stone, wood, metal (welding/casting), or clay/ceramics. Each requires vastly different tools.
* **Formal or Informal Training:** * **Academic:** A Bachelor of Fine Arts (BFA) provides a broad foundation and networking.
* **Apprenticeship:** Finding a master sculptor to study under is the traditional way to learn specialized techniques like bronze casting or stone carving.


* **Study Anatomy:** Even for abstract work, understanding the skeletal and muscular structure of humans and animals is essential for creating "weight" and balance.

## 2. Developing a Portfolio

You need a body of work to prove your capability to clients, galleries, and lenders.

* **Experimentation:** Spend time finding your "voice." Are you a minimalist, a realist, or a conceptual artist?
* **Documentation:** Invest in high-quality photography. Sculpture is 3D; you need photos from multiple angles with professional lighting to represent the work accurately.

---

## 3. Business and Legal Foundations

Transitioning from "artist" to "professional sculptor" requires a shift in mindset toward business management.

| Step | Action Item | Why it Matters |
| --- | --- | --- |
| **Business Plan** | Define your niche and pricing. | Ensures your workshop is financially sustainable. |
| **Legal Structure** | Register as a Sole Proprietor or LLC. | Protects your personal assets and handles taxes. |
| **Insurance** | Get "Artist's Insurance." | Covers equipment theft and liability if a heavy piece injures someone. |

---

## 4. Setting Up the Workshop (The Studio)

Sculpture is a heavy industry. Your workspace needs more than just a desk.

* **Ventilation and Safety:** Many sculpting processes involve dust, fumes, or heat. Proper HVAC and PPE (masks, goggles) are non-negotiable.
* **Load Bearing:** Sculptures are heavy. Ensure your floor can support the weight of stone or large metal stocks.
* **Specialized Equipment:** * **Subtractive:** Chisels, hammers, angle grinders, pneumatic tools.
* **Additive:** Kilns, welding machines, casting crucibles, armatures.



## 5. Marketing and Representation

Once the workshop is running, you need to find homes for your pieces.

* **The "Call for Entry":** Apply to juried shows and competitions to build your resume.
* **Gallery Representation:** Approach galleries that fit your style to handle sales on your behalf.
* **Public Commissions:** Look for "Percent for Art" programs where cities dedicate a portion of construction budgets to public sculpture.

---

* **Maintenance:** Learn how to preserve and clean your finished works so they last for centuries.
* **Networking:** Join organizations like the **International Sculpture Center** to stay updated on grants and exhibitions.

Format of the Recipe Pattern
To use this pattern, your prompt should make the following fundamental contextual statements:

I would like to achieve X 

I know that I need to perform steps A,B,C 

Provide a complete sequence of steps for me 

Fill in any missing steps 

(Optional) Identify any unnecessary steps

You will need to replace "X" with an appropriate task. You will then need to specify the steps A, B, C that you know need to be part of the recipe / complete plan.

Examples:

I would like to  purchase a house. I know that I need to perform steps make an offer and close on the house. Provide a complete sequence of steps for me. Fill in any missing steps.

I would like to drive to NYC from Nashville. I know that I want to go through Asheville, NC on the way and that I don't want to drive more than 300 miles per day. Provide a complete sequence of steps for me. Fill in any missing steps.
