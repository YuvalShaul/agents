## The Alternative Approaches pattern

The Alternative Approaches pattern is a structural communication and problem-solving strategy used to present multiple solutions to a single problem.  
Rather than advocating for a "one size fits all" answer, it focuses on evaluating different methods based on specific constraints, trade-offs, and contexts.

In software engineering, this is often referred to as the Strategy Pattern, while in business and design, it is a framework for Option Analysis.

### Prompt Structure
To get the most out of an AI when using the Alternative Approaches pattern, your prompt should explicitly ask for a comparison of different methods, a breakdown of trade-offs, and a specific recommendation based on a context you provide.

### Core Components of the Pattern
To use this pattern effectively, you typically break down the presentation into three distinct layers:
| Component | Purpose |
| The Context | Defines the problem statement and the goals that must be met.|
| The Candidates | A curated list of 2–4 distinct methods to solve the problem.|
| The Trade-off Matrix | A comparison of pros, cons, costs, and risks for each candidate.|


### Why Use It?
The primary goal of the Alternative Approaches pattern is to avoid "The Golden Hammer" bias—the tendency to use the same tool for every problem.
- Objective Decision Making: It shifts the conversation from "What do I like?" to "What fits the constraints?"
- Risk Mitigation: By exploring alternatives, you often uncover hidden pitfalls in your primary choice.
- Stakeholder Buy-in: Showing that you have considered multiple paths builds trust and demonstrates due diligence.

### Common Implementation: The Comparison Matrix
When applying this pattern, a comparison table is the most effective way to visualize the "Alt" paths.  
If you were choosing a data storage method, for example:  
**Case Study: Choosing a Database**
| Approach | Best For | Pros | Cons | 
|Relational (SQL) | Structured data, Finance | Data integrity, ACID compliance  | Harder to scale horizontally|
| Document (NoSQL) | Unstructured data, CMS | Flexible schema, High speed | Potential for data duplication|
| In-Memory (Redis) | Real-time analytics, Caching | Extremely low latency | Limited by RAM capacity|

## ## Comparison of Task Execution Approaches

If you are assigned a **Task X** (e.g., "Automate Customer Support Ticket Resolution"), here are the three primary ways to structure the technical approach:

### ### 1. Workflow-Based Automation (Deterministic)

This uses AI to interpret data, but follows a rigid, linear path (like n8n or Zapier with LLM steps).

* **Pros:** Highly predictable, easiest to audit/test, lower token costs.
* **Cons:** Fails on "out-of-bounds" requests; requires manual updates for every new edge case.

### ### 2. Single Autonomous Agent (Reasoning Loop)

A single LLM equipped with tools (Search, Database, API) that decides the steps to take in real-time.

* **Pros:** Flexible and handles complex, multi-step queries without pre-defined paths.
* **Cons:** Higher "hallucination" risk; can get stuck in infinite loops; harder to predict cost.

### ### 3. Multi-Agent System (Orchestrated)

Different agents are assigned specific roles (e.g., one agent "Researches," another "Drafts," a third "Reviews").

* **Pros:** Highest accuracy; mimics human departmental structures; compartmentalizes errors.
* **Cons:** Highest complexity and latency; most expensive to develop and run.

---

### ### Summary Table: Pros & Cons

| Approach | Reliability | Flexibility | Implementation Cost |
| --- | --- | --- | --- |
| **Workflow** | **High** | Low | Low to Medium |
| **Single Agent** | Medium | **High** | Medium |
| **Multi-Agent** | **High** | **Very High** | High |

> **Pro-Tip:** Start with the **Workflow** approach for 80% of client tasks. Only move up to **Multi-Agent** systems when the task requires high levels of nuance or collaborative reasoning.
