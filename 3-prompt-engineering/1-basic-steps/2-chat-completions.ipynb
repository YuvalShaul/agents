## Chat Completions API Demo

- This is the new Chat-Completions API.
- The name is a bit misleading, since there are no "chats" really, it is just a message, but is meant for chats.
- See documentation here: 
- The API endpoint is: https://api.openai.com/v1/chat/completions
- The idea is that you send a list of messages, and the model will generate a response.
- The messages can be of type "user" or "assistant".
- The messages can also have a "system" type, which is used to set the system prompt.
- The messages can also have a "name" field, which is used to identify the role of the message.
- This way, a system prompt cannot be overridden by a user message.

```python
import os
from dotenv import load_dotenv   # from the python-dotenv library - load the OpenAI API key from file
from openai import OpenAI
load_dotenv(dotenv_path='../.env')
os.environ["OPENAI_API_KEY"]

client = OpenAI()

completion = client.chat.completions.create(   # This is the main API call - It sends the request to OpenAI 
  model="gpt-4o-mini",
  messages=[
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": "Hello!"}
  ]
)
print(f'Num of msgs: {len(completion.choices)}')
print(f'msg: {completion.choices[0].message.content}')
print(f'role: {completion.choices[0].message.role}')
```
