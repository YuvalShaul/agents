### Exercise: Support LLM Dispatcher


Build an agent that uses an LLM to route a user to one of three departments: **Technical**, **Billing**, or **Legal**.

#### 1. The Setup

* **The State:** Define a simple `TypedDict` that holds the `user_input` and a `category`. (No reducers or message lists needed yet!)
* **The LLM Node:** Create a function where you pass the user's input to an LLM with a **System Prompt**.
* *The Instruction:* "Look at the user's request. Output exactly one word: 'technical', 'billing', or 'legal'."



#### 2. The Three Specialist Nodes

* Create three separate functions.
* Each function should print a unique message (e.g., "Connecting to a lawyer..." or "Opening a support ticket...").

#### 3. The Routing Logic

* Use a **Conditional Edge**.
* This edge should look at the `category` string the LLM produced and decide which of the three specialist nodes to visit next.

#### Test your agent

1. **Input A:** "I need to change my credit card on file." (Works great!)
2. **Input B:** "I'm having a problem with my thing." (**Fails!**)
