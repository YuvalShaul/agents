```python
from typing import TypedDict

class UserProfile(TypedDict):
    username: str
    age: int


# ERROR: Missing 'age'
user: UserProfile = {"username": "Alice"}   # This is an error, but you'll see it only if you have a linter

# ERROR: 'age' should be int, not str
user: UserProfile = {"username": "Alice", "age": "twenty-five"}     # Again - an error that will not cause an exception
```

```python
from typing import  Annotated, get_type_hints


texed_int =  Annotated[int, 'Some more info.'] 

number: texed_int = 10

print(number)
print(texed_int)
```
