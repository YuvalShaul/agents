# Exercise: Building your first RAG Agent

In the previous exercise (2-embeddings-exercise), you turned PDF documents into a **Vector Database** (a library of "numbers"). In this exercise, you will build an **Agent** that acts as the Librarian.

### The Objective

To create a "Thinking" loop where the AI realizes it doesn't know an answer, searches your Medcare Vector DB, and then answers the user.

---

### Step 1: Define the "Search" Tool

Before the Agent can think, it needs "hands."

* **Instruction:** Create a function that takes a string (a search query) and connects it to the **Chroma** or **FAISS** database you built in the previous lab.
* **Goal:** When the Agent "calls" this tool, the tool should return the top 3 most relevant text chunks from the Medcare PDFs.

### Step 2: The "System Prompt" (The Identity)

You must tell the Agent who it is and what tools it has.

* **Instruction:** Write a "System Message" that explains:
1. "You are a Medcare Policy Assistant."
2. "You have a tool called `search_knowledge_base`."
3. "If you don't know an answer, you **must** use the tool before saying you don't know."



### Step 3: Create the "ReAct" Loop

This is the heart of the agent. You need to set up a loop (using a library like LangGraph or a simple Python `while` loop).

* **Instruction:** Set up the logic so that the Agent follows this sequence:
* **Input:** Receive the User Question.
* **Thought:** Determine if the question is about Medcare.
* **Action:** Call the search tool.
* **Observation:** Read the text returned by the tool.
* **Final Answer:** Combine the tool's text into a helpful response.



### Step 4: Testing the "Search" Logic

Now, test the Agent with a question that **requires** the database.

* **Instruction:** Ask the agent: *"What is the annual review protocol for Virtual Consultation Platforms?"*
* **Verification:** Look at your `Role: TOOL` output.
* Did the agent actually call the tool?
* Did the tool return the paragraph about "annual reviews"?



---

### Step 5: Handling "Empty" Results (The Challenge)

What happens if the database doesn't have the answer?

* **Instruction:** Ask the agent: *"What is the protocol for hospital parking?"*
* **Observation:** Since your PDFs don't mention parking, the tool will return irrelevant text.
* **Goal:** Observe if your Agent is smart enough to say: *"I found some information on Medcare protocols, but nothing specifically about parking."*
