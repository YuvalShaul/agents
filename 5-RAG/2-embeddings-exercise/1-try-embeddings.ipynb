## PDF Chunking and Storage Exercise

Follow these steps to process, store, and query your documents.

---

### 1. Document Selection

* Download the [**"Attention Is All You Need"**](https://research.google/pubs/attention-is-all-you-need/) research paper (Transformer model).
* Download the [**"Leave No Trace"**](https://lnt.org/wp-content/uploads/2023/07/Leave-No-Trace-Skills-Ethics-Guide.pdf) outdoor ethics guide.

### 2. Chunking Instructions

* Use **Recursive Character Text Splitting** to chunk these PDF files.
* Set a **Maximum Chunk Size** of 500 characters.
* Set a **Chunk Overlap** of 50 characters to maintain context between snippets.
* Ensure the output format is a list of plain text strings.

---

### 3. Storing in Chroma

* Initialize a **Chroma DB** persistent client.
* Create a new collection named `student_knowledge_base`.
* Generate **Embeddings** for each chunk using a standard model (e.g., `all-MiniLM-L6-v2`).
* Add the text chunks and their corresponding embeddings to the collection.
* Include **Metadata** for each entry (e.g., `"source": "transformer_paper"`).

---

### 4. Querying Examples

Use the following natural language queries to test your database:

* **Query 1:** "What is the primary benefit of the Multi-Head Attention mechanism?"
* **Query 2:** "What are the specific guidelines for disposing of waste in the wilderness?"
